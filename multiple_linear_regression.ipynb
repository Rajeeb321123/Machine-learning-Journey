# Multiple Linear Regression

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('50_Startups.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```
dataset
```

## Encoding categorical data

>  Rememeber Dummy Variable Trap is taken cared by the class itself, we dont have to worry about it

> We also dont have worry about p-value and statistical signigicance as it will taken cared by the class from scikitlearn

```
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers=[("encoder", OneHotEncoder(),[3])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
X
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Training the Multiple Linear Regression model on the Training set

```
# Same as simple Regression, class will determine it is simple or multiple regression by itself
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## Predicting the Test set results

```
X_test[2]
```

```
y_pred = regressor.predict(X_test)
```

## Visualize the Prediction

```
np.shape(y_pred)
```

```
# To visualize the predicted y_pred and real value y_test
np.set_printoptions(precision=2) #precision =2 will only 2 decimal value, so we can easily visualize it
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_pred),1)), axis=1)) #reshape to vertical. axis = 1 means horizontal concatenation here
```

```
# y = b0( or intercept_) + b1*X1 + b2*X2 ..... b6X3
regressor.coef_, regressor.intercept_
```

## Backward Elimination: This is completely optional as new scikit library do it by themself

```
# Building the optimal model using Backward Elimination
import statsmodels.api as sm
X = np.append(arr = np.ones((50, 1)).astype(int), values = X, axis = 1)
X_opt = X[:, [0, 1, 2, 3, 4, 5]]
X_opt = X_opt.astype(np.float64)
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
regressor_OLS.summary()
X_opt = X[:, [0, 1, 3, 4, 5]]
X_opt = X_opt.astype(np.float64)
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
regressor_OLS.summary()
X_opt = X[:, [0, 3, 4, 5]]
X_opt = X_opt.astype(np.float64)
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
regressor_OLS.summary()
X_opt = X[:, [0, 3, 5]]
X_opt = X_opt.astype(np.float64)
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
regressor_OLS.summary()
X_opt = X[:, [0, 3]]
X_opt = X_opt.astype(np.float64)
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
regressor_OLS.summary()
```
